```python
q20 = df["Exam_Score"].quantile(0.20)
q50 = df["Exam_Score"].quantile(0.50)

def create_risk(score):
    if score <= q20:
        return "High Risk"
    elif score <= q50:
        return "Medium Risk"
    else:
        return "Low Risk"
```

```python
import pandas as pd
import pickle

model = pickle.load(
    open("../artifacts/model.pkl", "rb")
)

preprocessor = pickle.load(
    open("../artifacts/preprocessor.pkl", "rb")
)

feature_names = (
    preprocessor.get_feature_names_out()
)

coef_df = pd.DataFrame({
    "Feature": feature_names,
    "Coefficient": model.coef_[0]
})

coef_df = coef_df.sort_values(
    by="Coefficient"
)

print(coef_df)
```
